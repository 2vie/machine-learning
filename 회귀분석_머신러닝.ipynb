```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
import statsmodels.api as sm
```

```python
housing = fetch_california_housing()
df_X = pd.DataFrame(housing.data, columns=housing.feature_names)
df_y = pd.DataFrame(housing.target, columns=housing.target_names)
df = pd.concat([df_X, df_y], axis=1)
df.describe()
```

```python
y=df[housing.target_names] #
X=df[housing.feature_names]
X=sm.add_constant(X)
model = sm.OLS(y,X)
result_ols=model.fit()
result_ols.summary()
```

```python
df['MedHouseVal_int'] = df['MedHouseVal'].round().astype(int)
df['MedHouseVal_int'].describe()
```

```python
y = df['MedHouseVal_int']
X = df[housing.feature_names]
X = sm.add_constant(X)
model = sm.GLM(y,X, family=sm.families.Poisson())
result_poisson=model.fit()
result_poisson.summary()
```

```python
median_value = df['MedHouseVal'].median()
df['MedHouseVal_binary'] = (df['MedHouseVal'] > median_value).astype(int)
df['MedHouseVal_binary'].describe()
```

```python
y=df['MedHouseVal_binary']
X=df[housing.feature_names]
X = sm.add_constant(X)
model = sm.Logit(y,X)
result_logit = model.fit()
result_logit.summary()
```
